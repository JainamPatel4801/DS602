#### 1. Load the breast cancer dataset using datasets.load_breast_cancer(),
#### Use scatter plots to look at the data (use different colors for the classes), 
#### Use box plots to look at the data

```python
from sklearn.datasets import load_breast_cancer
import matplotlib.pyplot as plt

data = load_breast_cancer()
X = data.data
y = data.target

# here the scatter plot for the data....
colours = ['g' if label == 0 else 'y' for label in y]
plt.figure(figsize=(8, 5))
plt.scatter(X[:, 0], X[:, 1], c=colours)
plt.xlabel(data.feature_names[0])
plt.ylabel(data.feature_names[1])
plt.title('DATA in Scatter Plot')
plt.show()

# here the box plot for the data....
plt.figure(figsize=(8, 5))
plt.boxplot([X[y == 0][:, 0], X[y == 1][:, 0]], labels=data.target_names)
plt.ylabel(data.feature_names[0])
plt.title('DATA in Box Plot')
plt.show()
```

#### 2. Create a training and test set with random_state = 3

```python
from sklearn.model_selection import train_test_split


# here i train the data as ask for with the random state = 3
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=3)
```

#### 3. train a decision tree model on the training set ,train a random forest model on the training set

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier



# here i train the Decision Tree for the data....
dt_m = DecisionTreeClassifier(random_state=3)
dt_m.fit(X_train, y_train)



# here i trian the Logistic Regression for the data as asked for with the random state = 3....
lr_m = LogisticRegression(max_iter=10000, random_state=3)
lr_m.fit(X_train, y_train)



# here this code for the Random Forest for the data with the random state = 3....
rf_m = RandomForestClassifier(random_state=3)
rf_m.fit(X_train, y_train)
```

#### 4. Which model performed better?

```python
from sklearn.metrics import accuracy_score

y_pre_dt = dt_m.predict(X_test)
y_pre_lr = lr_m.predict(X_test)
y_pre_rf = rf_m.predict(X_test)

acc_dt = accuracy_score(y_test, y_pre_dt)
acc_lr = accuracy_score(y_test, y_pre_lr)
acc_rf = accuracy_score(y_test, y_pre_rf)



print(f"This is Decision Tree Accuracy: {acc_dt}")
print(f"This is Logistic Regression Accuracy: {acc_lr}")
print(f"This is Random Forest Accuracy: {acc_rf}")


# here i compare the model for the best perfomance selection between all of the models....
if acc_dt > acc_lr and acc_dt > acc_rf:
    best_m = "Decision Tree"
elif acc_lr > acc_dt and acc_lr > acc_rf:
    best_m = "LOGISTIC REGRESSION"
else:
    best_m = "RANDOM FOREST"

print(f"The BEST performing model in this data would be: {best_m}")
```

#### 5. Which features were most relevant in each model (a part from random forest)?

```python

dt_f_imp = dt_m.feature_importances_
lr_coef = lr_m.coef_[0]


# here i tried to get the main feature importance for the both decision tree and logistic regression.... 
print("The Importance for Decision Tree Feature :")
for feature, importance in zip(data.feature_names, dt_f_imp):
    print(f"{feature}: {importance}")

print("\n\n\nThe Importance and Coefficients for Logistic Regression :")
for feature, coefficient in zip(data.feature_names, lr_coef):
    print(f"{feature}: {coefficient}")
```

#### 6. Use grid search to improve performance on the best performed model

```python
from sklearn.model_selection import GridSearchCV

# here i used the grid search while fitting the models and with the random state =3....
p_grid = {
    'n_estimators': [100, 200, 300],
    'max_depth': [None, 10, 20, 30],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(RandomForestClassifier(random_state=3), p_grid, cv=5, n_jobs=-1)
grid_search.fit(X_train, y_train)

best_rf_m = grid_search.best_estimator_
best_rf_acc = accuracy_score(y_test, best_rf_m.predict(X_test))

print(f"The best RANDOM Forest model after adding grid search has ACCURACY of {best_rf_acc}")
```
