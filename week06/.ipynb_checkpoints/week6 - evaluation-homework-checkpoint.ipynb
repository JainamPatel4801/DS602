## Question 3
- Create a simple confusion matrix. Use `0.5` as the threshold for whether an observation should be considered True or False.  
- Comment on what you observe.

```python
# insert code here

# this is a pre set threshold for classification....
threshold = 0.5

# here i created a simple confusion matrix fro representation.....
true_pos = ((results['actual'] == 1.0) & (results['prob'] >= threshold)).sum()
false_pos = ((results['actual'] == 0.0) & (results['prob'] >= threshold)).sum()
true_neg = ((results['actual'] == 0.0) & (results['prob'] < threshold)).sum()
false_neg = ((results['actual'] == 1.0) & (results['prob'] < threshold)).sum()

confusion_matrix = pd.DataFrame({
    'Predicted TRUE': [true_pos, false_pos],
    'Predicted FALSE': [false_negative, true_neg]
}, index=['Actual TRUE', 'Actual FALSE'])

confusion_matrix
```

Insert comments here

- Here the confusion matrix gives us the idea of performance of the model with pre-defined threshold of 0.5....
- here are the values which is used 
    * TRUE POSITIVE : the model appropriately predicated True.
    * FALSE POSITIVE : the model inappropriately predicated True.
    * TRUE NEGATIVE : the model appropriately predicated False.
    * FALSE NEGATIVE : the model inappropriately predicated False.
- here i got an idea from predictions that here there are more true positive and false negative, showing that the model might have some what bias towards predicting the true class....    
    
    

## Question 4
- Calculate the accuracy, precision, and recall `without` using any pre-built functions.
- Which measure would you use if this was your model and why?

```python
# insert code here

# here i calculated accuracy with this fromula...
accuracy = (true_pos + true_neg) / len(results)

# here i calculated precision.....
precision = true_pos / (true_pos + false_pos)

# here i calculated recall....
recall = true_pos / (true_pos + false_neg)


print(f"the ACCURACY : {accuracy}")
print(f"the PRECISION : {precision}")
print(f"the RECALL : {recall}")
```

Insert comments here

**defination:**
- Accuracy: The proportion of accurately predicted outcomes, which is a measure of total model performance.
- Precision: The proportion of true positive predictions among all positive predictions, which is appropritate when false positives are costly.
- Recall: The proportion of true positive predictions among all actual positives, which is important when we are not having actual positives is costly.

here the choice measures to main or the specific problem and its consequences. If false positives or false negatives have different impacts on the sets, you might prioritize either precision or recall accordingly for ur model....

## Question 5
- Construct a ROC curve plot.  
- Comment on the fit of the model based on this chart.

```python
# Insert code here

# here we have to calculate the True Positive Rate and False Positive Rate firstly for different thresholds....
thresholds = np.linspace(0, 1, 100)

t_p_r = []
f_p_r = []

for th in thresholds:
    tp = ((results['actual'] == 1.0) & (results['prob'] >= th)).sum()
    fn = ((results['actual'] == 1.0) & (results['prob'] < th)).sum()
    tn = ((results['actual'] == 0.0) & (results['prob'] < th)).sum()
    fp = ((results['actual'] == 0.0) & (results['prob'] >= th)).sum()
    t_p_r.append(tp / (tp + fn))
    f_p_r.append(fp / (fp + tn))

# thhis is the ROC curve plot...
plt.plot(f_p_r, t_p_r)
plt.xlabel('FALSE Positive Rate')
plt.ylabel('TRUE Positive Rate')
plt.title('ROC CURVE PLOT')
plt.show()
```

Insert comments here


- here firstly the curve mainly focuses on the trade off between the true positive rate and false positive rate as the classification thresholds value changes. 
- moslty the good model which is perfectly fit would resluts in a cruve that is close to the y axis on the top.
- if the curve inclined to the left top corner which means the model perform well which is same in our case. the chosen threshold is may for the problem which based upon the model like recall, precision balancing etc...
