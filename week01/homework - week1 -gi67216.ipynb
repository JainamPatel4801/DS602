## Question 1. We need to know what's included in this data.

#### 1-A) How many countries are present?
**Approach** : Calculates the number of unique number of countries in the 'country' column of our DataFrame. The result is stored in the variable num_countries. For safe side we can convert the country column case so we can overcome the issue if there is same countries but it might having lowercase and uppercase difference. So we will leave that as it is.


````python
## Unique Countries Counting code...
num_countries = df['country'].astype(str).str.lower().nunique()
print(f"Number of COUNTRIES: {num_countries}")
````

#### 1-B) What's the timespan of the data?
**Approach** : For this we can use Min and Max function which will return minimum value from year column and maximum value from year column and from that we can find the timespan of data.

````python
start_year = df['year'].min()
end_year = df['year'].max()
print(f"Timespan of the DATA: {start_year} to {end_year}")
````

#### 1-C)Which countries are missing years?
**Approach** : First we will get the list of years from start year & end year and then we can find the missing_years for each country using the groupby with country. 

````python
## Used range function will give list of years from start year and end year ,in range end_year will be excluded so we have incremented by 1...
all_years = set(range(start_year, end_year + 1))
## We will use missing_years function which will find which are the years are missing if there is missing year for that country...
def missing_years(group):
    years_for_country = set(group['year'])
    return all_years - years_for_country
## We will use GroupBy by country to find the missing_years...
missing_data = df.groupby('country').apply(missing_years)


'''missing_data is series as index will be country and value will be set of missing_years for that country
    if there is no missing_years it will be empty set so converting empty set into boolean will result False otherwise True
'''

countries_missing_years = missing_data[missing_data.astype(bool)]
print("Countries which is having Missing Year in data : " ,list(countries_missing_years.index))
## Find the count of missing_years for that country...
missing_years_df = pd.DataFrame(countries_missing_years,columns=["missing_years"]).reset_index()
missing_years_df["missing_years_count"] = missing_years_df["missing_years"].apply(lambda row:len(row))
missing_years_df
````

#### 1-D) How many rows have at least 1 non-null values across the columns 3-57?
**Approach** : We can use slice indexing concept to check only that columns for that we have to check None value and after that notnull().any() will return us the True and False ,if all columns from 3-57 is having none value it returns False other wise True and then we can filter the dataframe and get only rows which is having at least one Not Null values from columns 3-57.

````python
non_null_rows = df[df.iloc[:, 4:58].notnull().any(axis=1)]
num_non_null_rows = len(non_null_rows)
print(f"Number of rows with atleast 1 NON-NULL value across columns 3-57: {num_non_null_rows}")
````

## Question 2: Wide data isn't usually helpful.
Convert this to a long format with the following columns:  
1. country  
2. year  
3. diagnosis method: mapping should be: 

```
  {'newsp':'Pulmonary Smear (+)'
  ,'newsn':'Pulmonary Smear (-)'
  ,'newep':'Extrapulmonary'
  ,'newrel':'Relapse'}   
 ```

4. gender: male or female.   
5. age: lower_age - higher_age, e.g., 0-14  
6. number of cases  

__Note__: You can use `replace` (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html)

#### Approch
**A**: Here we copy our original data frame to new data frame named as df2 and after that we will use rename function and other neccessary functions which i will describe in the comments accordingly.

````python
df2 = df.copy()
rename_dict = {
    'new_sp':'Pulmonary Smear (+)',
    'new_sn':'Pulmonary Smear (-)',
    'new_ep':'Extrapulmonary',
    'newrel':'Relapse'
}
# We can use rename to rename and replace the some part of column name by using replace method and for mapping we can use rename_dict...
df2 = df2.rename(columns={col: col.replace(key, rename_dict[key]) for col in df2.columns for key in rename_dict if key in col})

# We will Melt the dataframe to reshape using the melt function...
df_melted = df2.melt(id_vars=['country', 'year'], value_name='number_of_cases', var_name='combined_info')

# We will Extract the needed information from the combined_info...
df_melted['diagnosis_method'] = df_melted['combined_info'].str.extract(r'(^.*?)_')
df_melted['gender'] = df_melted['combined_info'].str.extract(r'_(.)\d{2,4}$').replace({'m': 'Male', 'f': 'Female'})
df_melted['age'] = df_melted['combined_info'].str.extract(r'(\d{2,4})$').replace({
    '014': '0-14', '1524': '15-24', '2534': '25-34', '3544': '35-44', '4554': '45-54', '5564': '55-64', '65': '65+'
})
# After that we will Drop the combined_info column as it's no longer needed...
df_melted.drop(columns=['combined_info'], inplace=True)

# We will use manual reorder for Reordering columns for our clarity...
df_melted = df_melted[['country', 'year', 'diagnosis_method', 'gender', 'age', 'number_of_cases']]

df_melted
````

## Question 3: Create a graph that shows when countries started to report TB cases.

````python
earliest_reported = df_melted.dropna(subset=['number_of_cases']).groupby('country')['year'].min().reset_index()
plt.figure(figsize=(30, 120))
sns.scatterplot(x='year', y='country', data=earliest_reported, color='blue', s=50)
for _, row in earliest_reported.iterrows():
    plt.axvline(x=row['year'], color='gray', linestyle='--', alpha=0.5)
    plt.axhline(y=row['country'], color='gray', linestyle='--', alpha=0.5)

plt.title('Year When Countries Started Reporting TB Cases')
plt.xlabel('Year')
plt.ylabel('Countries')
plt.grid(axis='x', linestyle='--')
plt.tight_layout()
plt.show()
````
