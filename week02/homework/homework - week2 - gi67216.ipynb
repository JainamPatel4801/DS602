## Question 4. Analyze the relationship between the wholeprice `mmr` and the `salesprice`.
- What is the overall correlation?  
- What happens to the correlation as you increase the maximum selling price window (e.g., 0-10000, 0-20000,...)?
- Create a scatter plot with the relationship.  
- Create a graph that shows how the correlation changes.

````python
correlation = df['mmr'].corr(df['sellingprice'])
print("Correlation between MMR and Saleprice : ",correlation)
````

````python
df["sellingprice"].max()
````

**<u>Sample narrative description.</u>**

As max value of selling price 23000 we will give window 10000,20000 and 30000 to find the how coorelationship changes with increasing the maximum price window

````python
windows = [10000, 20000,30000]
correlations = {}
for window in windows:
    filtered_df = df[df['sellingprice'] <= window]
    correlation = filtered_df['mmr'].corr(filtered_df['sellingprice'])
    correlations["0-"+str(window)] = correlation
print(correlations)
````

**<u>Sample narrative description.</u>**

If we change the maxmimum price window from 10000 to 20000 and 20000 to 30000 the coorelation between mmr and sellingprice increases

#### <u>  scatter plot with the relationship. </u> 

````python
plt.figure(figsize=(10, 6))
plt.scatter(df['mmr'], df['sellingprice'], alpha=0.5)
plt.title('Relationship between MMR and Selling Price')
plt.xlabel('MMR')
plt.ylabel('Selling Price')
plt.grid(True)
plt.show()
````

####      Graph That shows how the coorealtion changes with increase the maximum price window

````python
correlation_matrix = pd.DataFrame([correlations])
plt.figure(figsize=(10, 3))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', cbar_kws={'label': 'Correlation Coefficient'})
plt.title('Correlation between MMR and Selling Price for Different Windows')
plt.ylabel('Correlation Coefficient')
plt.show()
````

**<u>Sample narrative description.</u>**

Here we can see the correlation in the form of heat map which give the idea that with higher the number have more correlation.

## Question 5. How does the distribution of `salesprice` vary by make for cars made between 2010-2015?
- Use the 10 most common makes (your cleaned version of this column).  
- Create a graph that answers this question.

````python
## Get the 10 most common makes using value_counts()
common_makes = df['make'].value_counts().head(10)
common_makes_list = list(common_makes.index)
print("10 Most common Makes : ",common_makes_list)
````

**<u>Approach</u>**

we first get the top 10 most common Maker from the data frame and use that to make the graph required to show the selling price.  

````python
df2 = df[df["make"].isin(common_makes_list)]
filtered_df = df2[(df2['year'] >= 2010) & (df2['year'] <= 2015)]
plt.figure(figsize=(15, 8))
sns.boxplot(x='make', y='sellingprice', data=filtered_df)
plt.title('Distribution of Selling Price by Make (2010-2015)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
````

````python
avg_prices = filtered_df.groupby(['make', 'year'])['sellingprice'].mean().reset_index()
plt.figure(figsize=(15, 8))
sns.lineplot(data=avg_prices, x='year', y='sellingprice', hue='make')
plt.title('Average Selling Price by Make (2010-2015)')
plt.xlabel('Year')
plt.ylabel('Average Selling Price')
plt.grid(True)
plt.legend(loc='upper left')
plt.tight_layout()
plt.show()
````
